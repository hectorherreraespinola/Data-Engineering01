Considerando los tipos de datos existentes, donde no se encuentran valores, así como información separada por comas, se procederá a hacer un trabajo de ETL.

Comienzo generando una copia de los DF 

```python
Amazon_Titles_copy =Amazon_Titles.copy()
Disney_Titles_copy=Disney_Titles.copy()
Hulu_Titles_copy=Hulu_Titles.copy()
Netflix_Titles_copy= Netflix_Titles.copy()
```

```python
#Realizo el agregado de una columna para cada DF creado. De esa manera puedo asignarle la plataforma
Amazon_Titles_copy['Plataforma'] = 'Amazon'
Disney_Titles_copy['Plataforma'] = 'Disney'
Hulu_Titles_copy['Plataforma'] = 'Hulu'
Netflix_Titles_copy['Plataforma'] = 'Netflix'
```

```python
#Reemplao valores de campos vacíos en la columna duration solo para la información de Hulu y Netlix
Hulu_Titles_copy['duration'].fillna('Sin dato',inplace=True)
Netflix_Titles_copy['duration'].fillna('Sin dato',inplace=True)
```

```python
#Genero una función que me permita dividir la columna duration
def separa_columna_duration(DF):
    DF['temporadas']=DF["duration"].apply(lambda i: i.split()[0] if i.split()[1]=='Seasons' else '0')
    DF['duracion_peliculas']= DF['duration'].apply(lambda i: i.split()[0] if i.split(" ",1)[1]=='min' or i.split(" ",1)[1]=='mim'  else '0')
    return DF
```

```python
Amazon_Titles_copy =separa_columna_duration(Amazon_Titles_copy)
Disney_Titles_copy = separa_columna_duration(Disney_Titles_copy)
Hulu_Titles_copy = separa_columna_duration(Hulu_Titles_copy)
Netflix_Titles_copy = separa_columna_duration(Netflix_Titles_copy)
 
```

```python
#Elimino la columna duration en cada archivo 
Amazon_Titles_copy.drop(columns='duration',inplace=True)
Disney_Titles_copy.drop(columns='duration',inplace= True)
Hulu_Titles_copy.drop(columns= 'duration', inplace=True)
Netflix_Titles_copy.drop(columns='duration',inplace=True)
```

```python
#Chequeo la información
Amazon_Titles_copy.head()
```

```python
#Realizo la integración en un solo DF a fines practicos
peliculas_y_series=pd.concat([Amazon_Titles_copy,Disney_Titles_copy,Hulu_Titles_copy,Netflix_Titles_copy],axis=0)
```

```python
#chequeo el Dataframe unico
peliculas_y_series.head(10)
```

```python
#Elimino última columna adicional
peliculas_y_series.drop(columns='Unnamed: 0',inplace=True)
```

```python
#chequeo el Dataframe con la columna eliminada
peliculas_y_series.head(10)
```

```python
#Modifico las letras a mayuscula
peliculas_y_series.columns=['Show_Id','Type','Title','Director','Cast','Country','Date_Added','Release_Year','Rating', 'Listed_In','Description','Plataform','Seasons','Movie_Duration']
```

```python
#chequeo el Dataframe unico
peliculas_y_series.head(10)
```

```python
#Verifico la infrmación del DF creado
peliculas_y_series.info()
```

```python
#Transformar tipo de dato cantidad 
peliculas_y_series.Seasons=peliculas_y_series.Seasons.astype(int)
```

```python
#Transformo el tipo de dato de duración
peliculas_y_series.Movie_Duration=peliculas_y_series.Movie_Duration.astype(int)

```

```python
#Paso el archivo final a csv
peliculas_y_series.to_csv(r"..\Datasets\peliculas_y_series.csv", index=False, header=True)
```
